## Use List Comprehensions to Filter Data
#### Modify our JSON data by restricting it to only those entries that have a director and an IMDb link. We can do this with a list comprehension.  
#### syntax: [expression for element in source_list if filter_expression] - with conditional filter expression


```python
# used list comprehensions in the form to compress code that would have been done in a for loop.
# create a filter expression for only movies with a director and an IMDb link

wiki_movies = [movie for movie in wiki_movies_raw if ('Director' in movie or 'Directed by' in movie) and 'imdb_link' in movie]
```

```python
wiki_movies_df = pd.DataFrame(wiki_movies)
len(wiki_movies_df)
```

```python
wiki_movies_df.head()
```

```python
wiki_movies = [movie for movie in wiki_movies_raw
               if ('Director' in movie or 'Directed by'in movie)
                   and 'imdb_link' in movie
                   and 'No. of episodes' not in movie]
```

```python
# Step 3: After looping through every key, add the alternative titles dict to the movie object.
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
        if len(alt_titles) > 0:
            movie['alt_titles'] = alt_titles
        
        return movie
```
